```
# Import torch and other required modules
import torch
import numpy as np
```

```
#Function 1: torch.from_array() / tensor.numpy()
# Example 1
a = np.random.rand(3,4,5,6)
print(f'a: {type(a)}, {a.dtype}')
b = torch.from_numpy(a)
print(f'b: {type(b)}, {b.dtype}')
c = torch.tensor(a)
print(f'c: {type(c)}, {c.dtype}')
d = c.numpy()
print(f'd: {type(d)}, {d.dtype}')
```

```
# Example 2 
a1 = np.random.randint(low=0, high=20, size=(2,3,5))
print(f'a1: {type(a1)}, {a1.dtype}')
b1 = torch.from_numpy(a1)
print(f'b1: {type(b1)}, {b1.dtype}')
c1 = torch.tensor(a1)
print(f'c1: {type(c1)}, {c1.dtype}')
d1 = c1.numpy()
print(f'd1: {type(d1)}, {d1.dtype}')
```

```
#Function 2: torch.unsqueeze() / torch.squeeze()
# Example 1
print('b',b.shape,'\n',b)
_g = torch.unsqueeze(b, dim=1) 
_h = torch.squeeze(_g) 
#print(_g.shape, _h.shape)
print('_g\n',_g.shape,'\n',_g)
print('_h\n',_h.shape,'\n',_h)
```

```
# Example 2
print('b',b.shape,'\n',b)
_g2 = torch.unsqueeze(b, dim=2) 
_h2 = torch.squeeze(_g2) 
#print(_g.shape, _h.shape)
print('_g2\n',_g2.shape,'\n',_g2)
print('_h2\n',_h2.shape,'\n',_h2)
```

```
#tensor.transpose() / tensor.permute()
# Example 1
print('b',b.shape,'\n',b)
_e=b.permute((0,3,2,1))
_f=b.transpose(1,3)
print('_e\n',_e.shape,'\n',_e)
print('_f\n',_f.shape,'\n',_f)
```

```
# Example 2
print('b',b.shape,'\n',b)
_e2=b.permute((3,1,2,0))
_f2=b.transpose(0,3)
print('_e2\n',_e2.shape,'\n',_e2)
print('_f2\n',_f2.shape,'\n',_f2)
```
